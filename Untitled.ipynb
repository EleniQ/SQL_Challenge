```python
from sqlalchemy import create_engine
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import psycopg2
from config import password
```

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/SQL_Homework')
connection = engine.connect()
```

```python
salaries = pd.read_sql("SELECT * FROM \"salaries\"", connection)
salaries.min ()
salaries.max ()
salaries.head ()
```

```python
hi,plot=plt.subplots (figsize=(15, 10))
plt.hist(salaries["salary"], bins = 10, color = '#3d8a2c', edgecolor = 'black')
plt.title('Most Common Salaries of Employees', fontsize=14)
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.show()
```
